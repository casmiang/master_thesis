```python
def gen_random():
    L = 100
    x = np.random.rand() * L
    y = np.random.rand() * L
    z = np.random.rand() * L
    
    return [x,y,z]
```

```python
np.random.seed(0)

RC = []

for i in range(OC.shape[0]):
    RC.append(gen_random())
    
RC = np.array(RC)

np.savetxt( "random_catalogs/0.cat", RC)
```
