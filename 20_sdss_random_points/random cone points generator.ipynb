```python
def getPoint():
    u = np.random.rand()
    v = np.random.rand()
    theta = u * 2.0 * np.pi
    phi = np.arccos(2.0 * v - 1.0)
    r = (np.random.rand())**(1/3.)
    sinTheta = np.sin(theta)
    cosTheta = np.cos(theta)
    sinPhi = np.sin(phi)
    cosPhi = np.cos(phi)
    x = r * sinPhi * cosTheta
    y = r * sinPhi * sinTheta
    z = r * cosPhi

    return [x, y, z]
```

```python
points = []

for i in range(1000):
    points.append(getPoint())
    
points = np.array(points)
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')
ax.set_aspect('equal')


ax.scatter(points[:,0], points[:,1], points[:,2], s=10, alpha=0.2)
```

```python
def getPoint2(r_0, r_1, t_0, t_1, p_0, p_1):
    """
    t's & p's must be in degrees
    """
    
    t_0 = t_0 * np.pi / 180
    t_1 = t_1 * np.pi / 180
    
    u = np.random.rand()    
    theta = u * (t_1 - t_0 ) + t_0
    
    v_0 = np.cos((90-p_0)*np.pi/180)
    v_1 = np.cos((90-p_1)*np.pi/180)
    v = np.random.rand()
    phi = np.arccos( (v_0 - v_1)* v + v_1  )
    
    R = (1 - r_0/ r_1) * random.rand() + r_0/ r_1 
    r = r_1 * R **(1/3.)

    sinTheta = np.sin(theta)
    cosTheta = np.cos(theta)
    sinPhi = np.sin(phi)
    cosPhi = np.cos(phi)

    x = r * sinPhi * cosTheta
    y = r * sinPhi * sinTheta
    z = r * cosPhi

    return [x, y, z]
```

```python
points = []

R = 300
for i in range(10000):
    points.append(getPoint2(0,R,135,225,54,0))

points = np.array(points)

fig = plt.figure(figsize=(6,6))
ax = fig.gca(projection='3d')
ax.set_aspect('equal','box')
ax.set_zlim(-R, R)
ax.set_xlim(-R, R)
ax.set_ylim(-R, R)

ax.scatter(points[:,0], points[:,1], points[:,2], s=1, alpha=0.1)
```
