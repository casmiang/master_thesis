# cosmosim.org

SQL Query

Halo Catalog from the MDPL2 simulation, using the RockStar Halo Finder,  at redshift 0 (snapshot=85)
Halo Low Mass Limit = 1e13

```
SELECT x, y, z, Mvir  FROM MDPL2.Rockstar
WHERE snapnum=85 AND x<250 AND y<250 AND z<250 AND Mvir > 1e13
ORDER BY Mvir desc
```
